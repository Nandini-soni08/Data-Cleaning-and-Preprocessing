```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isna().any()
```

```python
df.dropna(thresh=2)
```

```python
df.fillna('Not Available',inplace = True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.capitalize()
```

```python
columns_to_capitalize = ['show_id',	'type',	'title',	'director',	'cast',	'country',	'date_added',	'release_year',	'rating',	'duration',	'listed_in','description']  # replace with your actual column names

for col in columns_to_capitalize:
    df[col] = df[col].astype(str).str.capitalize()
```

```python
df['Date_added'] = pd.to_datetime(df['Date_added'], errors='coerce')
```

```python
df
```

```python
df['Release_year'] = df['Release_year'].astype('int')
```
